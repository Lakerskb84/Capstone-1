```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
fish=pd.read_csv('fishing.csv')
print(fish)
```

```python
fish=fish.groupby(['Year','Species'])['CPUE','RPN','RPW'].mean()
print(fish)
```

```python
fish=fish.reset_index()
print(fish)
```

```python
catch=pd.read_csv('Fishing_results.csv')
print(catch)
```

```python
catch=catch[(catch.Year >1999) & (catch.Year <2018)]
print(catch)
```

```python
catch=catch.groupby(['Year','Species'])['Total Number Caught','Mean Weight (kg)'].mean()
print(catch)
```

```python
catch=catch.reset_index()
```

```python
m1=pd.merge(fish, catch)
print(m1)
```

```python
m1.replace(0.0, np.NaN)
```

```python
m1=m1.fillna(m1.mean())
print(m1)
```

```python
carbon=pd.read_csv('Carbon Data.csv')
print(carbon)
```

```python
carbon=carbon.rename(columns={'Time' : 'Year'})
print(carbon)
```

```python
carbon=carbon[(carbon.Year >1999) & (carbon.Year <2018)]
print(carbon)              
```

```python
carbon['co2 ppm']=carbon['co2 ppm'].apply(lambda x:float(x.split(' ')[0]))
print(carbon)
```

```python
m2=pd.merge(m1, carbon)
print(m2)
```

```python
from sklearn.preprocessing import LabelEncoder

le=LabelEncoder()
m3=m2['Species_a']= le.fit_transform(m2['Species'])
m3=m2[['Species_a','CPUE','RPN','RPW','Total Number Caught','Mean Weight (kg)','co2 ppm']]
print(m3)
```

```python
m3.info()
```

```python
from sklearn import preprocessing

s=m3[['CPUE','Total Number Caught','Mean Weight (kg)','co2 ppm']]
y=m3['Species_a']

standard_x= preprocessing.scale(s)
print(standard_x.shape)
```

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression
reg= LogisticRegression(multi_class='multinomial', solver='newton-cg')
rfe=RFE(reg, 4)
fit=rfe.fit(s, y)
print(fit.ranking_)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
X=standard_x
y=m3['Species_a']
params= {'C':[.001, .01, 1.0, 10.0]}
logreg=LogisticRegression(multi_class='multinomial', solver='newton-cg')

X_train, X_test, y_train, y_test= train_test_split(X, y, train_size=0.30, random_state=42)
logreg_cv= GridSearchCV(logreg, param_grid=params, cv=3)
logreg_cv.fit(X_train, y_train)

print('Logreg best parameter: {}'.format(logreg_cv.best_params_))
print('logreg accuracy:{}'.format(logreg_cv.best_score_))
```

```python
from sklearn import metrics

X_test,X_train,y_test,y_train= train_test_split(X, y, train_size=0.30, random_state=42)

logreg=LogisticRegression(C=10.0,multi_class='multinomial', solver='newton-cg')
logreg.fit(X_train, y_train)
tr=metrics.accuracy_score(y_train, logreg.predict(X_train))
test=metrics.accuracy_score(y_test, logreg.predict(X_test))
print(tr)
print(test)
```
