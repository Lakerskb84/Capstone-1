```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
fish=pd.read_csv('fishing.csv')
print(fish)
```

```python
c=pd.read_csv('sea_data.csv')
print(c)
```

```python
c['DateAnalyzed']=pd.to_datetime(c['DateAnalyzed'])
c['Year']=c['DateAnalyzed'].dt.year
print(c)
```

```python
c=c[['Year','Temp (C)','pH']]
print(c)
```

```python
c.fillna(c.mean())
```

```python
carbon=pd.read_csv('Carbon Data.csv')
print(carbon)
```

```python
carbon=carbon.rename(columns={'Time':'Year'})
print(carbon)
```

```python
m1=pd.merge(fish, c, on='Year', how='outer')
print(m1)
```

```python
m2=pd.merge(m1, carbon, on='Year', how='outer')
print(m2)
```

```python
m2.info()
```
