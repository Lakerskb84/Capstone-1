```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fish=pd.read_csv('fishing.csv')
print(fish)
```

```python
fish=fish.groupby(['Year','Species'])['CPUE','RPN','RPW'].mean()
print(fish)
```

```python
fish=fish.reset_index()
print(fish)
```

```python
catch=pd.read_csv('Fishing_results.csv')
print(catch)
```

```python
catch=catch[(catch.Year >1999) & (catch.Year <2018)]
print(catch)
```

```python
catch=catch.groupby(['Year','Species'])['Total Number Caught','Mean Weight (kg)'].mean()
print(catch)
```

```python
catch=catch.reset_index()
```

```python
m1=pd.merge(fish, catch)
print(m1)
```

```python
m1=m1.fillna(m1.mean())
print(m1)
```

```python
carbon=pd.read_csv('Carbon Data.csv')
print(carbon)
```

```python
carbon=carbon.rename(columns={'Time' : 'Year'})
print(carbon)
```

```python
carbon=carbon[(carbon.Year >1999) & (carbon.Year <2018)]
print(carbon)              
```

```python
carbon['co2 ppm']=carbon['co2 ppm'].apply(lambda x:float(x.split(' ')[0]))
print(carbon)
```

```python
sea=pd.read_csv('sea_data.csv')
print(sea)
```

```python
sea['DateAnalyzed']=pd.to_datetime(sea['DateAnalyzed'])
sea['Year']=sea['DateAnalyzed'].dt.year
print(sea)
```

```python
sea.info()
```

```python
sea=sea.groupby('Year', as_index=False)['Temp (C)','pH'].mean()
print(sea)
```

```python
sea=sea.fillna(sea.mean())
print(sea)
```

```python
m2=pd.merge(m1, carbon)
print(m2)
```

```python
m3=pd.merge(m2, sea, how='outer')
print(m3)
```

```python
plt.hist(m3['Total Number Caught'], bins=30)
plt.show()
```

```python
plt.hist(m3['CPUE'], bins=30)
plt.show()
```

```python
plt.bar(m3['Species'], m3['Total Number Caught'], color='g') 
plt.xticks(rotation=90)
plt.show()
```
