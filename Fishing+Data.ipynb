```python
#Impoting pandas,numpy and matplotlib.pyplot.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# loading fishing data as pandas DataFrame, fishing then printing.
fishing='fishing_results.csv'
fishing_df=pd.read_csv(fishing)
fishing_df
```

```python
#Dropping NaN values using dropna() and saving results as clean.
clean=fishing_df.dropna()
print(clean)
```

```python
# Using all Bering Sea data after the year 1987. Saved results as clean1 and printed
clean1=clean[(clean.Area== 'Bering Sea') & (clean.Year >1987)]
print(clean1)
```

```python
#Broke down dataframe clean1 into 4 colums,'Year','Area','Species' and 'Mean Weight (kg)
clean1[['Year', 'Area', 'Species', 'Mean Weight (kg)']]
```
